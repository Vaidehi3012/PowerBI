● Power BI Desktop

  - Power BI is a Microsoft business intelligence (BI) tool used to connect, transform, analyze, and visualize data to generate interactive reports and dashboards for better decision-making.

  - Key Components of Power BI Ecosystem
1. Power BI Desktop

Free Windows application used for data modeling, DAX calculations, and report creation

Supports Power Query for data transformation

Used by analysts and developers

Example: Creating sales dashboards using the Global Superstore dataset.

2. Power BI Service

Cloud-based platform (powerbi.com)

Used for publishing, sharing, scheduling refreshes, and collaboration

Enables dashboards and app creation

Example: Sharing a sales dashboard with management.

3. Power BI Mobile

Mobile app for Android & iOS

Allows real-time access to dashboards

Optimized visuals for mobile viewing

Example: Sales managers tracking KPIs on mobile devices.

4. Power BI Gateway

Acts as a bridge between on-premise data sources and Power BI Service

Required for scheduled refresh from local databases

Example: Connecting on-prem SQL Server sales data to Power BI Service.

2. Compare the following Power BI visuals:
● Pie Chart vs Donut Chart
● Bar Chart vs Column Chart
When would you prefer one over the other? Give one example for each pair.

  - 1. Pie Chart vs Donut Chart
  - Pie Chart

Displays data as slices of a full circle to show part-to-whole relationships

Simple and easy to understand

Best when there are very few categories (2–4)

When to prefer a Pie Chart:

When you want a quick, simple comparison of proportions

When exact values are not critical

Example:
Showing sales contribution by region (East, West, Central, South)

  - Donut Chart

Similar to a pie chart but with a hole in the center

Allows placement of KPIs or total values in the center

More visually appealing and modern

When to prefer a Donut Chart:

When you want to display percentages along with a total value

When dashboard aesthetics matter

  - Example:
Showing percentage of total sales by product category with total sales in the center

2. Bar Chart vs Column Chart
  - Bar Chart

Uses horizontal bars

Easier to read long category names

Ideal for comparing many categories

When to prefer a Bar Chart:

When category labels are long

When comparing values across multiple product types

Example:
Comparing total sales by product sub-category

  - Column Chart

Uses vertical bars

Best for time-based trends

Commonly used for performance comparisons over periods

When to prefer a Column Chart:

When analyzing changes over time

When categories are short or represent dates

  - Example:
Displaying monthly sales performance

3. Explain the significance of:
● Star schema vs Snowflake schema
● Primary key vs Foreign key in relationships (Power BI)
Why is cardinality important?

  - 1. Star Schema vs Snowflake Schema
  - Star Schema

Consists of a central fact table connected directly to dimension tables

Dimension tables are denormalized

Simple structure, easy to understand

Better performance in Power BI due to fewer joins

Significance / When to use:

Preferred for reporting and analytics

Faster query performance

Easier DAX calculations

Example:
Sales Fact table connected to Customer, Product, Region, and Date tables.

  - Snowflake Schema

Dimension tables are normalized into multiple related tables

More complex structure

Requires more joins, which can reduce performance

Significance / When to use:

Useful when data storage optimization is needed

Maintains data consistency in large databases

Less commonly used in Power BI models

  - Key Difference Summary
Star Schema	Snowflake Schema
Simple	Complex
Denormalized	Normalized
High performance	Lower performance
Recommended for Power BI	Not ideal for reports
2. Primary Key vs Foreign Key (in Power BI Relationships)
  - Primary Key

A column with unique values and no duplicates

Identifies each record in a table

Located in dimension tables

Example:
CustomerID in Customer table

  - Foreign Key

A column that references the primary key in another table

Allows tables to be linked together

Located in fact tables

Example:
CustomerID in Sales table referencing Customer table

3. Why Cardinality Is Important in Power BI

Cardinality defines the type of relationship between two tables:

One-to-One (1:1)

One-to-Many (1:*)

Many-to-Many (:)

  - Significance of Cardinality

Determines how filters flow between tables

Impacts accuracy of aggregations

Affects performance and report correctness

Best Practice:

Use One-to-Many relationships (Dimension → Fact)

Avoid Many-to-Many unless absolutely necessary

Example:

One Customer → Many Orders (1:M)

Why It Matters

Incorrect cardinality can cause:

Duplicate values

Wrong totals

Incorrect filtering in visuals

4. Differentiate between:
● Calculated column vs Measure
Also, define Row context and Filter context with simple examples.

  - Calculated Column vs Measure
Calculated Column

Created using DAX

Calculated row by row and stored in the table

Uses row context

Increases data model size

Value is fixed once calculated (changes only on data refresh)

Example (Calculated Column):
Total Price = Sales[Quantity] * Sales[Unit Price]

  - Measure

Created using DAX

Calculated at query time

Uses filter context

Does not increase model size

Value changes dynamically based on filters, slicers, and visuals

Example (Measure):
Total Sales = SUM(Sales[Total Price])


  - | Calculated Column   | Measure               |
| ------------------- | --------------------- |
| Stored in table     | Calculated on the fly |
| Uses row context    | Uses filter context   |
| Static values       | Dynamic values        |
| Increases file size | Efficient             |

  - Row Context

Definition:
Row context refers to the current row being evaluated during a calculation.

Used in:
Calculated columns

Simple Example:
If a Sales table has Quantity and Price columns, Power BI calculates each row separately:
Line Total = Sales[Quantity] * Sales[Price]

  - Filter Context

Definition:
Filter context refers to filters applied from visuals, slicers, rows, columns, or DAX functions.

Used in:
Measures

Simple Example:
Total Sales = SUM(Sales[Sales Amount])
Shows total sales

If a slicer selects Region = South, the measure recalculates sales only for the South region.

5. What is the difference between a report and a dashboard in Power BI?

  - Difference Between a Report and a Dashboard in Power BI
  - Report

A report is a collection of multiple pages of visuals

Created using Power BI Desktop

Uses data from one dataset

Allows detailed analysis and interaction (filters, slicers, drill-down)

Example:
A sales report with pages for regional sales, product performance, and monthly trends.

  - Dashboard

A dashboard is a single-page canvas

Created in Power BI Service (online)

Can combine visuals from multiple reports and datasets

Provides a high-level summary of key metrics

Less interactive than reports

Example:
An executive dashboard showing total sales, profit, and top-performing regions.
| Report             | Dashboard           |
| ------------------ | ------------------- |
| Multiple pages     | Single page         |
| Built in Desktop   | Built in Service    |
| Detailed analysis  | High-level overview |
| One dataset        | Multiple datasets   |
| Highly interactive | Limited interaction |
